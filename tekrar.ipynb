```python
import pandas as pd
from io import StringIO
import numpy as np

data= """
Date,Product,Category,Quantity,Price
2025-01-01,Phone,Electronics,2,1200
2025-01-02,Laptop,Electronics,3,800
2025-01-03,Headphones,Accessories,5,150
2025-01-04,Keyboard,Accessories,4,100
2025-01-01,Monitor,Electronics,1,700



"""


df=pd.read_csv(StringIO(data))
df



```

```python
groupSum=df.groupby("Category")["Quantity"].sum()
groupSum
```

```python
groupSum=groupSum.reset_index()
groupSum
```

```python
groupStats=df.groupby("Category")[["Quantity","Price"]].agg({
    "Quantity":"sum",
    "Price":"mean"
})
groupStats
```

```python
groupStats=groupStats.reset_index()
groupStats
```

```python
df["Cumulative Sum"]=df["Quantity"].cumsum()
df
```

```python
df["Cumulative Mean"]=df["Price"].expanding().mean()
df
```

```python
df["Rolling Window"]=df["Price"].rolling(window=2, min_periods=1).mean()
df
```

```python
df["Rank"]=df["Price"].rank(ascending=False)
df
```

```python
df["Category Rank"]=df.groupby("Category")["Price"].rank(ascending=False)
df
```
